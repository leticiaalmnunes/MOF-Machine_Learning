## 2. Comparao e discusso

Abaixo, obteremos um grfico para facilitar a visualizao dos resultados obtidos nos Cadernos `3.0`, `3.1`, `3.2` e `3.3`.

#### Definindo variveis

```python
modelos = ['Baseline', 'k-NN s/ otim.', 'k-NN c/ otim.', 'rvore s/ otim.', 'rvore c/ otim.', 'Floresta s/ otim.', 'Floresta c/ otim.']
rmses = [64.2888, 59.9279, 61.5296, 70.5299, 63.5425, 61.9264, 61.0360]
minimo=min(rmses)

bar_labels = ['Baseline', 'Regresso k-NN', '_Regresso k-NN', 'rvore', '_rvore', 'Floresta', '_Floresta']
bar_colors = ['darkgoldenrod', 'crimson', 'crimson', 'teal', 'teal', 'darkslateblue', 'darkslateblue']
```

#### Configurando o grfico de RMSE por modelo

```python
fig, ax = plt.subplots(layout='constrained', dpi=1200, figsize = (12, 8))

ax.bar(modelos, rmses, color=bar_colors, label=bar_labels)
ax.axhline(64.2888, color='tan', linestyle='--', label='Limite superior')
#ax.axhline(minimo, color='gray', linestyle='--', alpha=0.5) #linha de mnimo
ax.set_title("""RMSE  Modelo \n""",  fontsize = 20)
ax.set_xlabel("""Modelo""", fontsize = 15)
ax.set_ylabel("""RMSE (h) \n""", fontsize = 15)
ax.legend(title = 'Modelo', bbox_to_anchor=[1,1], fontsize = 'x-small', title_fontsize = 'small')

plt.xticks(rotation = -15)
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.savefig('RMSE_x_Modelo.png')
plt.show();
```

Abaixo, obteremos algumas estatsticas descritivas do target para conseguirmos fazer uma avaliao mais objetiva dos resultados observados no grfico acima.

```python
df=pd.read_csv("RAC_train.csv")
tempo = list(df['time'])
 
print(f'Maior tempo:{max(tempo)} h.')
print(f'Menor tempo:{min(tempo)} h.')
print(f'Mdia do tempo:{round(np.mean(tempo), 2)} h.')
print(f'Desvio padro do tempo:{round(np.std(tempo), 2)} h.')
print()
print(f'RMSE do baseline: {round(rmses[0], 2)} h')
print(f'Melhor RMSE: {round(min(rmses[1:]), 2)} h')
print(f'Pior RMSE: {round(max(rmses[1:]), 2)} h')
print(f'Mdia do RMSE: {round(np.mean(rmses[1:]), 2)} h')
print(f'Desvio padro do RMSE: {round(np.std(rmses[1:]), 2)} h')
```

Como podemos observar no grfico, o melhor modelo foi o **Regressor $k$-NN** sem a otimizao dos hiperparmetros, cujos atributos foram selecionados por **VIF** (Fator de Inflao da Varincia) e o RMSE foi de 59,93 h.

Tal resultado pode ser considerado estranho, levando em considerao que o $k$-NN  o modelo mais simples entre o trs sendo analisados e que a otimizao dos hiperparmetros no melhorou a performance. No entanto, como todos os modelos resultaram em RMSE's prximos, e parecidos com o do baseline, acreditamos que as pequenas mudanas na performance, evidenciadas pelo desvio padro de 3,5 h, so efeito do acaso. Ou seja, nenhum modelo apresentou uma performance suficientemente boa.
