## 2. Comparação e discussão

Abaixo, obteremos um gráfico para facilitar a visualização dos resultados obtidos nos Cadernos `3.0`, `3.1`, `3.2` e `3.3`.

#### Definindo variáveis

```python
modelos = ['Baseline', 'k-NN s/ otim.', 'k-NN c/ otim.', 'Árvore s/ otim.', 'Árvore c/ otim.', 'Floresta s/ otim.', 'Floresta c/ otim.']
rmses = [64.2888, 59.9279, 61.5296, 70.5299, 63.5425, 61.9264, 61.0360]
minimo=min(rmses)

bar_labels = ['Baseline', 'Regressão k-NN', '_Regressão k-NN', 'Árvore', '_Árvore', 'Floresta', '_Floresta']
bar_colors = ['darkgoldenrod', 'crimson', 'crimson', 'teal', 'teal', 'darkslateblue', 'darkslateblue']
```

#### Configurando o gráfico de RMSE por modelo

```python
fig, ax = plt.subplots(layout='constrained', dpi=1200, figsize = (12, 8))

ax.bar(modelos, rmses, color=bar_colors, label=bar_labels)
ax.axhline(64.2888, color='tan', linestyle='--', label='Limite superior')
#ax.axhline(minimo, color='gray', linestyle='--', alpha=0.5) #linha de mínimo
ax.set_title("""RMSE × Modelo \n""",  fontsize = 20)
ax.set_xlabel("""Modelo""", fontsize = 15)
ax.set_ylabel("""RMSE (h) \n""", fontsize = 15)
ax.legend(title = 'Modelo', bbox_to_anchor=[1,1], fontsize = 'x-small', title_fontsize = 'small')

plt.xticks(rotation = -15)
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.savefig('RMSE_x_Modelo.png')
plt.show();
```

Abaixo, obteremos algumas estatísticas descritivas do target para conseguirmos fazer uma avaliação mais objetiva dos resultados observados no gráfico acima.

```python
df=pd.read_csv("RAC_train.csv")
tempo = list(df['time'])
 
print(f'Maior tempo:{max(tempo)} h.')
print(f'Menor tempo:{min(tempo)} h.')
print(f'Média do tempo:{round(np.mean(tempo), 2)} h.')
print(f'Desvio padrão do tempo:{round(np.std(tempo), 2)} h.')
print()
print(f'RMSE do baseline: {round(rmses[0], 2)} h')
print(f'Melhor RMSE: {round(min(rmses[1:]), 2)} h')
print(f'Pior RMSE: {round(max(rmses[1:]), 2)} h')
print(f'Média do RMSE: {round(np.mean(rmses[1:]), 2)} h')
print(f'Desvio padrão do RMSE: {round(np.std(rmses[1:]), 2)} h')
```

Como podemos observar no gráfico, o melhor modelo foi o **Regressor $k$-NN** sem a otimização dos hiperparâmetros, cujos atributos foram selecionados por **VIF** (Fator de Inflação da Variância) e o RMSE foi de 59,93 h.

Tal resultado pode ser considerado estranho, levando em consideração que o $k$-NN é o modelo mais simples entre o três sendo analisados e que a otimização dos hiperparâmetros não melhorou a performance. No entanto, como todos os modelos resultaram em RMSE's próximos, e parecidos com o do baseline, acreditamos que as pequenas mudanças na performance, evidenciadas pelo desvio padrão de 3,5 h, são efeito do acaso. Ou seja, nenhum modelo apresentou uma performance suficientemente boa.
